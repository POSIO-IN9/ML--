```python
X_train, X_test, y_train, y_test = train_test_split(iris['data'],iris['target'],random_state=42)
```

## 학습 및 모델

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

## 평가 및 검증

```python
y_pred =knn.predict(X_test)
y_pred
```

```python
knn.score(X_test,y_test)
#1.0은 오버핏
```
